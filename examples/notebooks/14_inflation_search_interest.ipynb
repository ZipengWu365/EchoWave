# Inflation + search interest -> regime shift report

A beginner economics example: show how a pair of macro-adjacent signals can be summarized before you choose a model.

```python
import pandas as pd
from tsontology import profile_dataset
df = pd.read_csv('../data/inflation_search_interest.csv')
report = profile_dataset(df[['inflation','search_interest']].to_numpy(), domain='economics')
print(report.to_summary_card_markdown())
```
