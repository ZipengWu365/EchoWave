# Wearable recovery cohort

Inspect heterogeneity and adherence patterns in a longitudinal cohort.

```python
from tsontology import starter_dataset, profile_dataset

case = starter_dataset("wearable_recovery_cohort")
profile = profile_dataset(case["values"], domain="wearable", timestamps=case["timestamps"], channel_names=case["channels"])
print(profile.to_narrative_report())
```
