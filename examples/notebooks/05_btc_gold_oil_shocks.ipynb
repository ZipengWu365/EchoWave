# BTC / gold / oil shocks


A market comparison example aimed at rolling similarity and regime-sensitive interpretation.

```
from tsontology import starter_dataset, compare_series, rolling_similarity
case = starter_dataset("btc_gold_oil_shocks")
report = compare_series(case["btc"], case["gold"], left_name="BTC", right_name="Gold")
print(report.to_narrative_report())
rolling_similarity(case["btc"], case["gold"], window=30, step=5)[:3]
```
