# Air quality vs heatwave

Cross-domain demo: compare environmental signals under a heatwave-like drift.

```python
import pandas as pd
from tsontology import explain_similarity
df = pd.read_csv('../data/air_quality_heatwave.csv')
print(explain_similarity(df['temperature'].to_numpy(), df['pm25'].to_numpy(), left_name='temperature', right_name='pm25'))
```
