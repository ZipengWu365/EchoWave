# Industrial sensor network coupling

Cross-domain demo: show coupling and heterogeneity across multiple sensors.

```python
import pandas as pd
from tsontology import profile_dataset
df = pd.read_csv('../data/industrial_sensor_network.csv')
report = profile_dataset(df[['sensor_a','sensor_b','sensor_c']].to_numpy(), domain='industrial')
print(report.to_summary_card_markdown())
```
