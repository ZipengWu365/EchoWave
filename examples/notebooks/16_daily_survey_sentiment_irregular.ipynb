# Daily survey sentiment -> irregularity and burstiness

A beginner social-science example using explicit timestamps.

```python
import pandas as pd
from tsontology import profile_dataset
df = pd.read_csv('../data/daily_survey_sentiment_irregular.csv')
report = profile_dataset(df['sentiment'].to_numpy(), timestamps=df['timestamp'].to_numpy(), domain='social-science')
print(report.to_summary_card_markdown())
```
