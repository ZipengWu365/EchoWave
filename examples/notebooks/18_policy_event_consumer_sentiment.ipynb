# Policy event vs consumer sentiment

Cross-domain demo: look at structural change around a policy-like event.

```python
import pandas as pd
from tsontology import profile_dataset
df = pd.read_csv('../data/policy_event_consumer_sentiment.csv')
report = profile_dataset(df[['consumer_sentiment','retail_spending']].to_numpy(), domain='economics')
print(report.to_narrative_report())
```
