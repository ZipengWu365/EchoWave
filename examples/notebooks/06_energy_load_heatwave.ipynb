# Energy load vs heatwave

Profile a small multivariate energy dataset and render a shareable HTML report.

```python
from tsontology import starter_dataset, profile_dataset

case = starter_dataset("energy_load_heatwave")
profile = profile_dataset(case["values"], domain="energy", timestamps=case["timestamps"], channel_names=case["channels"])
print(profile.to_summary_card_markdown())
```

```python
Path("../outputs/energy_load_heatwave_report.html").write_text(profile.to_html_report(audience="operations"), encoding="utf-8") if False else None
```
