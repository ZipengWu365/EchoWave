# Sine vs noise


This is the smallest possible onboarding example. It shows why a rhythmic signal and pure noise should receive different structure reports.

```
import numpy as np
from tsontology import starter_dataset, profile_dataset
case = starter_dataset("sine_vs_noise")
sine = profile_dataset(case["left"])
noise = profile_dataset(case["right"])
print(sine.to_summary_card_markdown())
print(noise.to_summary_card_markdown())
```
