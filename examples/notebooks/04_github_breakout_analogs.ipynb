# GitHub breakout analogs


A social-friendly demo: compare a new repository trajectory to short-hype and durable-breakout analogs.

```
from tsontology import starter_dataset, compare_series, rolling_similarity
case = starter_dataset("github_breakout_analogs")
report = compare_series(case["target"], case["durable_breakout"], left_name="new_repo", right_name="durable_breakout")
print(report.to_summary_card_markdown())
windows = rolling_similarity(case["target"], case["durable_breakout"], window=24, step=4)
windows[:3]
```
