# Weekly website traffic


A product-style example: trend, weekly rhythm, and bursts in sessions and signups.

```
from tsontology import starter_dataset, profile_dataset
case = starter_dataset("weekly_website_traffic")
profile = profile_dataset(case["values"], domain="traffic", timestamps=case["timestamps"], channel_names=case["channels"])
print(profile.to_summary_card_markdown())
open("weekly_website_traffic_report.html", "w", encoding="utf-8").write(profile.to_html_report())
```
